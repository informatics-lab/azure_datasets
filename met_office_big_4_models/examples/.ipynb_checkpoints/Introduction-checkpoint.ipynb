# Met Office atmospheric models data 

The UK Met Office has made data from it's four world class weather models available for research purposes. 

These are made available under a Creative Commons licence see [LICENCE.txt](https://metdatasa.blob.core.windows.net/models/LICENCE.txt) for further details.

This notebook demonstrates some ways of working with the data, for more information about the models and datasets see the [README](https://metdatasa.blob.core.windows.net/models/README.html).



## A little background about our data

Our weather forecast models predict many weather parameters, such as air temperature, humidity or rainfall. These are predicted for every point on a regular grid, that is, as a function of latitude, longitude, and (when appropriate) altitude.

We regularly run "ensemble forecasts" and two of the datasets (MOGREPS-UK and MOGREPS-G) are ensemble datasets. An ensemble means that instead of starting one model run now to predict tomorrow, we start many, purturbing the intial parameters to give several possible futures (or realisations). The spread of these different realisations gives an indication of the uncertainty of the forecast, and can be very powerful information.

![An ensamble forecast](http://data.informaticslab.co.uk/images/ensemble_forecasting_schematic.png)

Each forecast point also has several measures of time. Firstly, time indicates the real world time that the data is forecast for. Secondly, reference_time is the time that the forecast was started from i.e. the 0th time point. Finally, forecast_period is the difference between the two i.e. the number of hours that have elapsed since the start of the forecast.


## Download a file

All files are stored with deterministic filenames at a http endpoint. For details on this see the [README](https://metdatasa.blob.core.windows.net/models/README.html).

Given this it's simple to create a function to determine the url for a specific data file and then download it.

```python
ROOT_URL = "https://metdatasa.blob.core.windows.net/models/"
```

```python
import datetime

def data_url(model_start, forecast_period_hrs=0, model="mo-atmospheric-ukv", 
             diagnostic_name="temperature_at_screen_level", ):
    """Determine url for the datafile."""
    
    model_start = model_start.replace(microsecond=0, second=0, minute=0)
    validity_time = model_start + datetime.timedelta(hours=forecast_period_hrs)
    key = f"{model}/{model_start:%Y%m%dT%H%M}Z/{validity_time:%Y%m%dT%H%M}Z-PT{forecast_period_hrs:04}H00M-{diagnostic_name}.nc"
    
    return ROOT_URL + key

data_url = data_url(datetime.datetime.now() - datetime.timedelta(hours=48))
data_url
```

```python
import urllib.request

file, responce = urllib.request.urlretrieve(data_url, data_url.split('/')[-1])
file
```

The file is now on our local machine.

## Loading Data

This data is in NetCDF format, which is a type of the widely used HDF format. You can use any compatible tools, but we recommend the Met Office's Iris Python module. First of all, let's make sure we've got Iris installed using Conda.

```python
%conda install -y -c conda-forge iris
```

Now let's load the file we just downloaded.

```python
import iris
data_cube = iris.load_cube(file)

data_cube
```

So here's the metadata for the data in our chosen file. We can see a lot of information here.

The data cube shape is (3 x 970 x 1042)
There are three numeric dimension coordinates (realization, projection_y_coordinate, projection_x_coordinate)
There are four coordinates that are not linked to a dimension (forecast_period, forecast_reference_time, height, time).
We can drill into the cube in various ways:

```python
data_cube.coord('time')
```

And here's a subset of the underlying data as a numpy array:

```python
data_cube.data[:10,:10:,:10]
```

You can slice a cube by index to extract particular pieces of data:

```python
# The first realization in the dataset
data_cube[0]
```

```python
# A specific grid point
data_cube[:,100,100]
```

## Plotting Data
Iris also has built-in functions to make plotting the cube easier:



```python
%matplotlib inline

import matplotlib.pyplot as plt
import iris.quickplot as qplt

# Set the size of the plot
plt.figure(figsize=(20, 20))

# Select the first realization
first_realization = data_cube[0]

# iris quickplot sets up our axes and labels so we don't have to
qplt.pcolormesh(first_realization, cmap='viridis')
```
